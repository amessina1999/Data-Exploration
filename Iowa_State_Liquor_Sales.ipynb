```python
# 1. Show the first five rows of you DF
df_iowa = pd.read_csv("iowa.zip")
df_iowa.head()
```

```python
#1. Change Column Names
df_iowa.columns = df_iowa.columns.str.replace(' ', '_')
df_iowa.rename(columns = {'Bottle_Volume_(ml)':'Volume_ml'}, inplace = True)
df_iowa.rename(columns = {'Sale_(Dollars)':'Sales'}, inplace = True)
df_iowa.rename(columns = {'Volume_Sold_(Liters)':'Liters_Sold'}, inplace = True)
df_iowa.rename(columns = {'Volume_Sold_(Gallons)':'Gallons_Sold'}, inplace = True)
df_iowa.rename(columns = {'Invoice/Item_Number': 'Invoice'}, inplace = True)
df_iowa.columns
```

```python
#1. Show data types
df_iowa.info()
```

```python
#1. Change data types

# Convert Date to DateTime
df_iowa['Date'] = df_iowa.Date.astype('datetime64')
# Convert Zip_Code to String
df_iowa.Zip_Code = df_iowa.Zip_Code.astype('str').str.split('.').str[0]
#Convert Country_Number to string
df_iowa.County_Number = df_iowa.County_Number.astype('str').str.split('.').str[0]
#Convert Vendor_Number to string 
df_iowa.Vendor_Number = df_iowa.Vendor_Number.astype('str').str.split('.').str[0]
#Convert Item_Number to string
df_iowa.Item_Number = df_iowa.Item_Number.astype('str')
#Convert Category to string
df_iowa.Category = df_iowa.Category.astype('str')

df_iowa.info()
```

```python
#1. show how many rows and columns 
df_iowa.shape
```

```python
#2. Clean the data show the total number of null values for each column 
df_iowa.isnull().sum()
```

```python
#2. Drop one of the columns based on null values
df_iowa = df_iowa.drop(['Store_Location'], axis = 1)
df_iowa.isnull().sum()
```

```python
#2. Drop Rows with null values 
df_iowa = df_iowa.dropna(subset = ['Address', 'Vendor_Number'])
df_iowa.isnull().sum()
```

```python
#3. Make sure there are no duplicates in the Invoice/Item Number column
df_iowa.Invoice.duplicated().sum()
# Invoice/Item_Number can be used as a primary key, because there are no duplicates
```

```python
# 3. Split the Invoice column to only recieve the invoice number
df_iowa.Invoice = df_iowa.Invoice.str.split('-').str[1]
df_iowa
```

```python
#4. Make a new column that combines Address, City, IA, and Zip Code 
df_iowa['mail_address'] = df_iowa.Address + ', ' + df_iowa.City + ', ' +  'IA, ' + df_iowa.Zip_Code
df_iowa
```

First we wanted to get insight about the vendors that are supplying the product. We can see that the top 5 suppliers are Diageo Americas, Sazerac Comapny Inc., Jim Beam Brands, Heaven Hill Brands, and Luxco Inc. 

```python
df_iowa.groupby('Vendor_Name')['Sales'].sum().sort_values(ascending = False)
```

Next we wanted to get an idea of how much sales are atributed to these suppliers. It seems to be the same top 5 suppliers with the exception fo Pernod Ricard USA replacing Luxco for 5th place. Although Luxco sells more units, Pernod Ricard has a higher value of sales in dollars.

```python
retail_mean = df_iowa.groupby('Item_Description').State_Bottle_Retail.mean()
cost_mean = df_iowa.groupby('Item_Description').State_Bottle_Cost.mean()
bottles_sold = df_iowa.groupby('Item_Description').Bottles_Sold.sum()
((retail_mean - cost_mean)*bottles_sold).sort_values(ascending = False)
```

Here we attempted to find the profit of each specific liquor, but there seemed to be a discrepancy in our findings. 
We took the mean of the retail cost subtracted by the mean of the selling price and multiplied it by the total units sold. 
The problem is that we got a higher value for profit than the actual amount of sales. Futher analysis was needed to find the error in this calculation.

```python
df_iowa[df_iowa.Item_Description == 'Fireball Cinnamon Whiskey'].State_Bottle_Cost.value_counts()
```

```python
df_iowa[df_iowa.Item_Description == 'Fireball Cinnamon Whiskey'].State_Bottle_Retail.value_counts()
```

By taking a closer look at the data, we realized that the grouping of Item_Description was not enough to be a determinate to calculate profit. We used Fireball as a test Item to see what the differant price points could be. At first we were under the impression that the stores could provide discounts from vendors for buying in bulk or could incentivise consumers to buy a certain products with deals. When we saw the results of the value_counts function, it seemed like there was too much of a spread in price. At this moment we realized that differant items can come in differant sizes. We went back to the data frame to look for other columns to dig deeper into. 

```python
df_iowa[df_iowa.Item_Description == 'Fireball Cinnamon Whiskey']['Volume_ml'].value_counts()
```

We decided to take a look at the Bottle_Volume_(mL) column while still filtering our search with fireball. We determind that there were multiple differant sized containers which would explain the spread in price. 

```python
df_iowa[df_iowa.Item_Description == 'Fireball Cinnamon Whiskey'].groupby(['Item_Description','Volume_ml']).State_Bottle_Retail.value_counts()
```

Here we expanded our grouping to Bottle_Volume_(ml) alongside Item_Description. We are now able to get a view of the variables that will give us a more accurate calculation. We decided to see if we could get a dataframe with more columns to describe the data. 

```python
group = df_iowa.groupby(['Item_Description','Volume_ml','State_Bottle_Cost', 'State_Bottle_Retail', 'Vendor_Name'])
retail= group.State_Bottle_Retail.mean()
cost= group.State_Bottle_Cost.mean()
bottles_sold = group.Bottles_Sold.sum()
df_profit = ((retail - cost)*bottles_sold).sort_values(ascending = False).reset_index()
df_profit
```

By grouping Item_Description, Bottle_Volume_(mL), State_Bottle_Cost, State_Bottle_Retail, Vendor_Name we were able to show all variables that lead to the desired outcome, which is profit per Item. The issue is the name of the column calculated, which is currently named 0. We need to rename this column. 

```python
df_profit.rename(columns = {0:'Profit'}, inplace = True)
df_profit
```

In this line of code we renamed the calculated column of 0 to Proft, which is a more accurate name for the new column. We then assingend this data frame to a new name variable name df_profit to preform more calculations

```python
df_profit.groupby('Item_Description').Profit.sum().sort_values(ascending = False)
```

We can see that Titos is the most profitable product.
